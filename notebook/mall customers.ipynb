```python
###piechart visualisation
labels=['FEMALE','MALE']
size=dataset['Gender'].value_counts()
explode=[0,0.5]
colors=['Gray','Cyan']
fig,ax=plt.subplots()
plt.rcParams['figure.figsize']=(5,5)
plt.pie(size,labels=labels,explode=explode,colors=colors,shadow=True,autopct='%1.1f%%')
plt.savefig("artifacts/piechart.pdf",format='pdf')
plt.show()
```

```python
###find out what age group has the higher income
age = dataset['Age']
annual_in=dataset['Annual_Income_(k$)']



plt.hist(age,annual_in,orientation='horizontal')

plt.savefig("images/hist.png")

plt.show()
```

```python
import seaborn as sns
age = dataset['Age']
annual_in=dataset['Annual_Income_(k$)']
plt.figure(figsize=(16,4))



sns.boxenplot(x=age,y=annual_in)
plt.title('Age and income')
plt.grid(True)
plt.savefig("artifacts/boxenplot.pdf",format='pdf')

plt.show()
```
